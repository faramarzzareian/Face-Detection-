```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import cvlib as cv
from cvlib.object_detection import draw_bbox
```

```python
im= cv2.imread(r'1.jpg' )
```

```python
plt.imshow(im[:,:,::-1])
```

```python
box,conf = cv.detect_face(im)
```

```python
conf
#confidence of detected objects . 
```

```python
labels = []
for i in range (len(conf)):
    labels.append('person')     #we add labels to the box which is append to each detected face with title of 'people'
```

```python
img=draw_bbox(im,box,labels,conf)
plt.imshow(im[:,:,::-1])
```
